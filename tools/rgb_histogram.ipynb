```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_rgb_histogram(image_path, title, save_path=None):
    # 이미지를 읽어들임
    image = cv2.imread(image_path)

    # 이미지를 RGB로 변환
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # 히스토그램 계산
    hist_red = cv2.calcHist([image_rgb], [2], None, [256], [0, 256])
    hist_green = cv2.calcHist([image_rgb], [1], None, [256], [0, 256])
    hist_blue = cv2.calcHist([image_rgb], [0], None, [256], [0, 256])
    
    mean_red = np.mean(image_rgb[:,:,0])
    mean_green = np.mean(image_rgb[:,:,1])
    mean_blue = np.mean(image_rgb[:,:,2])
    
    print("Mean Red: ", mean_red)
    print("Mean Green: ", mean_green)
    print("Mean Blue: ", mean_blue)

    # 그래프 그리기
    plt.figure(figsize=(6, 5))
    plt.plot(hist_red, color='red', label='Red', linewidth=2)
    plt.plot(hist_green, color='green', label='Green', linewidth=2)
    plt.plot(hist_blue, color='blue', label='Blue', linewidth=2)
    plt.title(title, fontsize=15, fontweight='bold')
    plt.xlabel('Intensity', fontsize=15, fontweight='bold')
    plt.ylabel('Frequency', fontsize=15, fontweight='bold')
    plt.ylim(0, 1700)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.legend(fontsize=15)
    plt.grid(True)
    if save_path:  # save_path가 제공되면 그래프를 해당 경로에 jpg 파일로 저장
        plt.savefig(save_path, format='jpg', bbox_inches='tight')
        print(save_path+" 저장 완료")
    plt.show()

```

```python
# 이미지 파일 경로
image_path = "image/auto_intg/sensor_add/preprocess_no_cut/20240313/20240313_071520.jpg"
# 함수 호출
plot_rgb_histogram(image_path, "RGB Histogram (7:15)","rgb_hist_20240313_0715.jpg")
```

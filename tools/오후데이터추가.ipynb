```python
import pandas as pd

# 엑셀 파일 경로 및 시트 이름 지정
file_path = "data/train_data_20240307-20240402_20240405_sun_info.xlsx"
sheet_name = "Sheet1"

# 데이터 불러오기
df = pd.read_excel(file_path, sheet_name=sheet_name,engine='openpyxl')
```

```python
# datetime 열 선택
datetime_column = df["datetime"]

# 데이터 유형 확인
print(datetime_column.dtype)
```

```python
# Extract the time component from the datetime column
df['time_only'] = df['datetime'].dt.time

# Define the afternoon start time
afternoon_start_time = pd.to_datetime("12:00:00").time()

# Select only the afternoon data
afternoon_data = df[df['time_only'] >= afternoon_start_time]

# Drop the temporary 'time_only' column if you don't need it anymore
afternoon_data.drop(columns=['time_only'], inplace=True)
```

```python
# 콘솔 출력
print(afternoon_data)

# CSV 파일로 저장
afternoon_data.to_excel("afternoon_data.xlsx", index=False)
```

```python
import shutil
import os

# Load the Excel data into a DataFrame
df = pd.read_excel('afternoon_data.xlsx',engine='openpyxl')  # Replace 'afternoon_data.xlsx' with your Excel file

# Specify the source folder where the photos are located
source_folder = r'image\auto_intg\sensor_add_change_position\preprocess_no_cut\train_image_20240405'  # Replace 'source_folder' with the path to your source folder

# Specify the destination folder where you want to move the photos
destination_folder = r'image\auto_intg\sensor_add_change_position\preprocess_no_cut\train_image_20240405_afternoon'  # Replace 'destination_folder' with the path to your destination folder

# Iterate over the rows of the DataFrame
for index, row in df.iterrows():
    # Get the filename from the 'filename' column
    filename = row['filename']
    
    # Construct the full path to the source photo
    source_path = os.path.join(source_folder, filename)
    
    # Construct the full path to the destination photo
    destination_path = os.path.join(destination_folder, filename)
    
    # Copy the photo from the source folder to the destination folder
    shutil.copy(source_path, destination_path)
    
    print(f"Photo '{filename}' copied to '{destination_folder}'")

print("All photos copied successfully.")
```
